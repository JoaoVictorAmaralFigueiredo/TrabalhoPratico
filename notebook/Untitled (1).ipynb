```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
df=pd.read_csv("incendios2000.csv", header=0)
df.head()
```

```python
df=df.sort_values(by=["causa","reacendimento"])
df.head(5)

## Tabela que prioriza os dados que contenham a causa e reacentimento
```

```python
df.sort_values(by=["area_total"])
df.head(5)

## Tabela que coloca em ordem (crescente) a area total  queimada
```

```python
df.groupby(["freguesia"]).freguesia.count().sort_values()[-10:].plot(kind="bar")

## Tabela que informa a quantidade de incendios por freguesia
```

```python
df.groupby(["concelho"]).concelho.count().sort_values()[-10:].plot(kind="bar")

## Tabela que informa a quantidade de incendios por concelho
```

```python
df.groupby(["distrito"]).distrito.count().sort_values()[-10:].plot(kind="bar")

## Tabela que informa a quantidade de incendios por distrito
```
